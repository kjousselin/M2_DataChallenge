# III - Recherche d'un modèle de ML efficace
On construit plusieurs modèles avec des fonctions d'activations et configurations différentes, puis on test sur nos données de validation à l'aide d'une fonction et de boucles.

```python
# Entrée : de 5912 à 24414   23095  1125


from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, Activation
from keras.layers import Conv1D, MaxPooling1D, GlobalAveragePooling1D, Flatten


import time

from google.colab import files

#taille_entree = df_train.shape[1]

def un_model_sequentiel(first_relu = False, FONCT_ACT = 'relu', avec_conv = False, mon_batch = 32, df = df_scale, y_true = Y_train_1_0, verbose = 0, end = False, sup = True):

    
    taille_entree = df.shape[1]
    
    model = Sequential()
    

    if avec_conv:
      # Add 1D convolutional layer with 32 filters, kernel size of 3 and 'relu' activation
      model.add(Conv1D(32, 3, input_shape=(taille_entree, 1))) #, activation='relu'))
      # Add a batch normalization layer
      model.add(BatchNormalization())
      # Add a max pooling layer
      model.add(MaxPooling1D(pool_size=2))
      # Add a dropout layer to reduce overfitting
      model.add(Dropout(0.5))

      model.add(Flatten())

      if first_relu:
          model.add(Dense(512, activation='relu'))
      else:
          model.add(Dense(512, activation=FONCT_ACT))
      
    else:
        if first_relu:
          model.add(Dense(512, input_shape=(taille_entree,), activation='relu'))
        else:
          model.add(Dense(512, input_shape=(taille_entree,), activation=FONCT_ACT))

    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    
    if sup:     # Une couche de neurones supplémentaire
        model.add(Dense(512, activation=FONCT_ACT))
        model.add(BatchNormalization())
        model.add(Dropout(0.5))

    model.add(Dense(256, activation=FONCT_ACT))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(128, activation=FONCT_ACT))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(64, activation=FONCT_ACT))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    if end:
        model.add(Dense(32, activation='relu'))
    else:
        model.add(Dense(32, activation=FONCT_ACT))

    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    model.fit(df, y_true, epochs=20, batch_size = mon_batch, verbose = verbose)

    return(model)

```

```python
def mesure_pred(model, df = df_scale_valid, y_true = Y_valid_1_0, verbose = 0):

  n = df.shape[0]

  prediction = model.predict(df, verbose = verbose)
  taux = prediction/y_true.reshape(n,1)*100

  # ecart moyen MAE
  MAE = np.sum(np.abs(prediction-y_true.reshape(-1,1)))/n
  MSE = np.sum(np.abs(prediction-y_true.reshape(-1,1))**2)/n

  t1=80
  t2=140
  print(f'\t\t\t\tDe {t1}% à {t2}% : {np.sum((t1<taux).reshape(n,)*(taux<t2).reshape(n,))} sur {n}')

  # xx valeurs entre t1 % et t2 % de la bonne valeur sur n !!
  nb_entre_deux_bornes = np.sum((t1<taux).reshape(n,)*(taux<t2).reshape(n,))

  pourc_ok = round(nb_entre_deux_bornes/n*100,1)
  
  return(MAE, MSE, pourc_ok)

```

```python
liste_fonct_activation = ['softmax', 'relu', 'tanh', 'softplus', 'selu', 'elu']

# Après quelques tests, j'ai enlevé les fonctions ci-dessous car elles paraissent moins efficace ici : 
# exponential sigmoid softmax softsign
```

```python

df_resultats = pd.DataFrame(columns=['fonc_act' , 'first', 'Conv', 'end', 'sup', 'MAE', 'MSE', 'pourc_ok' ])

for fonc in liste_fonct_activation:

    t0 = time.time()

    print(f"\n\nFonction d'activation : {fonc}")

    for first in [True, False]:   

        print(f"\n\tLa première est RELU : {first}")

        for Conv in [False, True]:    #[True, False]
            #Conv = False


            t00 = time.time()

            print(f"\t\tOn commence par une convolution : {Conv}")

            for sup in [False, True]:
                
                print(f'\t\t\tSup : {sup}')
                
                for end in [False, True]:
                    
                    print(f'\t\t\t\tend : {end}')

                    model = un_model_sequentiel(first_relu=first, FONCT_ACT = fonc, avec_conv=Conv, mon_batch=32, df = df_scale, y_true = Y_train_1_0, end=end, sup=sup)

                    res = mesure_pred(model, df = df_scale_valid, y_true = Y_valid_1_0)

                    print(f"\t\t\t\t\t\t(MAE, MSE, pourc_ok) = {res}")

                    df_temp = pd.DataFrame([(fonc, first, Conv, end, sup) + res], columns = ['fonc_act' , 'first', 'Conv', 'end', 'sup', 'MAE', 'MSE', 'pourc_ok' ])

                    df_resultats = df_resultats.append(df_temp, ignore_index=True)           

                    print(f"\t\t\t\t\t\tTemps pour ce modèle : {round((time.time()-t00)//60, 2)} min {round((time.time()-t00)%60, 0)} secondes.")

    # Enregistrement temporaire :
    df_resultats.to_csv(f'Resultats_de_valid_1_0_d_{fonc}.csv', encoding = 'utf-8-sig')
    files.download(f'Resultats_de_valid_1_0_d_{fonc}.csv')

    print(f"Temps pour cette fonction : {round((time.time()-t0)//60, 2)} min {round((time.time()-t0)%60, 0)} secondes.")


df_resultats.to_csv('Resultats_de_valid_1_0_d.csv', encoding = 'utf-8-sig')

files.download('Resultats_de_valid_1_0_d.csv')
```
